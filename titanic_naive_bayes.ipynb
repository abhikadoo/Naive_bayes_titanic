```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.metrics as met
from sklearn.model_selection import train_test_split
import sklearn.naive_bayes as nb
```

```python
dataset=pd.read_csv("titanic.csv")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(['Cabin'],axis=1, inplace=True)
```

```python
dataset['Embarked'].value_counts()
```

```python
embarked={'Embarked':{"S":0,"C":1,"Q":2}}
dataset.replace(embarked,inplace=True)
```

```python
dataset.head()
```

```python
dataset.dropna(inplace=True,how="any")
```

```python
dataset['Fare']=dataset['Fare'].astype(int)
dataset['Age']=dataset['Age'].astype(int)
```

```python
dataset=dataset.drop(['PassengerId','Name','Ticket'],axis=1)
```

```python
sex={"Sex":{'male':0,'female':1}}
dataset.replace(sex,inplace=True)
```

```python
dataset.head()
```

```python
features=dataset[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
label=dataset['Survived']
X_train,X_test,Y_train,Y_test=train_test_split(features,label,test_size=0.25,random_state=0)
```

```python
gnb=nb.GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
Y_pred=gnb.predict(X_test)
accuracy=met.accuracy_score(Y_test,Y_pred)
precision=met.precision_score(Y_test,Y_pred)
print('Accuracy=',accuracy,'Precision=',precision)
```
